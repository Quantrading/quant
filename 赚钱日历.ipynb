```python
import pandas as pd
import matplotlib.pyplot as plt
import quantstats as qs
from trader_tool.index_data import index_data
from trader_tool import jsl_data
from trader_tool.jsl_bond_data import jsl_bond_data
import matplotlib.pyplot as plt
import seaborn as sns
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
bond_data=jsl_bond_data()
bond=bond_data.bond_cb_index_jsl()
bond['date']=bond['price_dt']
bond['涨跌幅']=bond['price'].pct_change()*100
df=bond
df
```

```python
#月收益分布
df[-48:].plot.bar(x='date',y='涨跌幅',title='可转债等权指数',figsize=(12,6))
plt.show()
```

```python
#最近一年的情况
df[-12:].plot.bar(x='date',y='涨跌幅',title='可转债等权指数最近一年的情况',figsize=(12,4))
plt.show()
```

```python
#最近二年的情况
df[-24:].plot.bar(x='date',y='涨跌幅',title='可转债等权指数最近二年的情况',figsize=(12,4))
plt.show()
```

```python
#每个月平均收益绘制
df['标记']=df['date'].apply(lambda x:str(x).split('-')[1])
df1=df.groupby(by='标记').mean()
df1['标记']=df1.index
df1
```

```python
#相关性,收益和月份的相关性
df1['标记']=pd.to_numeric(df1['标记'])
sns.regplot(data=df1,x='标记',y='涨跌幅')
plt.show()
```

```python
#每一个月的赚钱情况
df1.plot.bar(x='标记',y='涨跌幅',title='每一个月的赚钱情况10年',figsize=(12,6))
plt.show()
```
